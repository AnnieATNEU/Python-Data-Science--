```python
import numpy as np
import pandas as pd
df = pd.read_csv (r'C:\Users\a_ade\Documents\Fall2021\CSC276\weather_data.csv')
df
```

# replacing list with single value

```python
new_df = df.replace(to_replace=['-99999', -88888],value= 0)
new_df
```

# replacing per column


```python
new_df= new_df.replace({
    'temperature': 0,
    'windspeed': 0,
    'event': '0'
}, np.nan)
new_df
```

# replacing by using mapping

```python
new_df= new_df.replace({
    -99999: np.nan,
    'NaN': 'Sunny',
})
new_df
```

# regex

```python
new_df = new_df.replace({'temperature':'[A-Za-z]', 'windspeed': '[a-z]'}, '', regex= True)
new_df
```
